#identifying unnecessary columns and removing it.
Columns having constant value and having no meaning in the dataset are unnecessary columns

```python
Unecessary_columns=['EmployeeCount','StandardHours','Over18']
df.drop(columns=Unecessary_columns,inplace=True,axis=1)
df
```

#Renaming the columns

```python
df.rename(columns={
    'Attrition':'EmployeeAttrition',
    'BusinessTravel':'TravelFrequency',
    'DailyRate':'StandardizedDailyWage',
    'MonthlyRate':'StandardizedMonthlyWage',
    'Education':'EducationLevel',
    'EnvironmentSatisfaction':'EnvironmentSatisfactionLevel',
    'RelationshipSatisfaction':'RelationshipSatisfactionLevel',
    'TrainingTimesLastYear':'TrainingCountLastYear',
    'WorkLifeBalance':'WorkLifeBalanceLevel',
    'JobInvolvement':'JobInvolvement',
    'HourlyRate':'StandardizedHourlyWage',
    'JobSatisfaction':'JobSatisfactionLevel'
    
},inplace=True)
```
